```python
import seaborn as sns
file = open("output/output.csv","r")
```

```python
dados = file.readlines()
```

```python
predict_function = lambda x: list(map(int, x[:-1].split(',')))[0]
y_pred = list(map(predict_function,dados))
```

```python
true_function = lambda x: list(map(int,x[:-1].split(',')))[1]
y_true= list(map(true_function,dados))
```

```python
acuracia  = 0
acumulador_acuracia = []

for indice, (v_true, v_pred) in enumerate(zip(y_true, y_pred)):
    booleano = 0
    if v_true == v_pred:
        booleano = 1

    acuracia = (acuracia* indice + booleano)/(indice+1)
    acumulador_acuracia.append(acuracia)
acuracia
```

```python
acumulador = [] 
somador = 0
ultimo_valor = None
for indice, (v_true, v_pred) in enumerate(zip(y_true, y_pred)):
    if v_true  == v_pred:
        somador+=1
    
    if indice % 2000 == 1999:
        ultimo_valor = somador/2000
        # acumulador.append(somador/ 1000)
        somador = 0 
    acumulador.append(ultimo_valor)
acumulador.append(  somador / (len(y_true)%2000 -1  ))
```
